### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
sum_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
sum_rides_by_type
```

```python
# 2. Get the total drivers for each city type
sum_driver_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
sum_driver_by_type
```

```python
#  3. Get the total amount of fares for each city type
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 

average_fare_per_ride = sum_fares_by_type / sum_rides_by_type
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = sum_fares_by_type / sum_driver_by_type
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
summary_df = pd.DataFrame({
    "Total Rides":sum_rides_by_type, 
    "Total Drivers":sum_driver_by_type, 
    "Total Fares":sum_fares_by_type,
    "Average Fare per Ride": average_fare_per_ride,
    "Average Fare per Driver": average_fare_per_driver})
summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
summary_df["Total Fares"] = summary_df["Total Fares"].map("${:,.2f}".format)
summary_df["Average Fare per Ride"] = summary_df["Average Fare per Ride"].map("${:,.2f}".format)
summary_df["Average Fare per Driver"] = summary_df["Average Fare per Driver"].map("${:,.2f}".format)
summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_and_dates_df = pyber_data_df.groupby(["date","type"]).sum()["fare"]
fares_and_dates_df
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fares_and_dates_df = fares_and_dates_df.reset_index()
fares_and_dates_df
```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fares_and_dates_df = fares_and_dates_df.pivot(index = "date", columns = "type", values = "fare")
fares_and_dates_df
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the givin dates, '2018-01-01':'2018-04-29'.
fares_and_dates_df = fares_and_dates_df.loc['2019-01-01':'2019-04-29']
fares_and_dates_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
fares_and_dates_df.index = pd.to_datetime(fares_and_dates_df.index)
```

```python
# 14. Check that the datatype for the index is datetime using df.info()
fares_and_dates_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_and_dates_df = fares_and_dates_df.resample("W").sum()
weekly_fares_and_dates_df
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.dates as mdates
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
months = mdates.MonthLocator()

x_axis = weekly_fares_and_dates_df.index
y_axis = weekly_fares_and_dates_df.values

fig, ax = plt.subplots(figsize=(10, 4))
ax.set_title("Total Fare by City Type")
ax.set_ylabel('Fare ($)')
ax.set_xlabel('Month')
ax.xaxis.set_major_locator(months)
ax.plot(x_axis, y_axis)
ax.legend(["Rural","Suburban","Urban"], loc=2)

plt.savefig("Total Fare by City Type")
```
